# For multiple faces with function

### import

```python
import cv2
import matplotlib.pyplot as plt
import time
```

### converttoRGB func.

```python
def convertToRGB(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### function for detecting faces

```python
def detect_faces(f_cascade,colored_img,scaleFactor=1.1):
    img_copy=colored_img.copy()
    gray = cv2.cvtColor(img_copy,cv2.COLOR_BGR2GRAY)
    faces=f_cascade.detectMultiScale(gray,scaleFactor=scaleFactor,minNeighbors=5);
    for (x,y,w,h) in faces:
        cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,255,0),2)
    return img_copy
```

### main coding

```python
test2=cv2.imread('/home/siva/Downloads/babies.jpg')
haar_face_cascade = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml')
lbp_face_cascade = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/lbpcascades/lbpcascade_frontalface.xml')
faces_detected_img=detect_faces(lbp_face_cascade,test2)
plt.imshow(convertToRGB(faces_detected_img))
```

# Face Detection and Recognition from webcam: 

# FACE DETECTION 

# face and eye detection in video capturing

```python
import cv2
import os
import numpy as np
face_cascade = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml')
eye_cascade = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/haarcascades/haarcascade_eye.xml')
cap = cv2.VideoCapture(0)
while(cap.isOpened()):
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,235,0),2)
    cv2.imshow('img', img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
if cap.isOpened():
    cap.release()
    cv2.destroyAllWindows()
```

###  generating yml file

```python
import cv2, os
import numpy as np
from PIL import Image
recognizer = cv2.face.LBPHFaceRecognizer_create()
detector = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml')
def getImagesAndLabels(path):
    imagePaths = [os.path.join(path,f) for f in os.listdir(path)]
    faceSamples=[]
    Ids=[]
    for imagePath in imagePaths:
        if(os.path.split(imagePath)[-1].split(".")[-1]!='jpg'):
            continue
        pilImage = Image.open(imagePath).convert('L')
        imageNp=np.array(pilImage)
        Id=int(os.path.split(imagePath)[-1].split(".")[1])
        faces = detector.detectMultiScale(imageNp)
        for (x,y,w,h) in faces:
            faceSamples.append(imageNp[y:y+h,x:x+w])
            Ids.append(Id)
    return faceSamples, Ids

faces, Ids = getImagesAndLabels('/home/siva/anaconda3/envs/env/Projects/imagedb')
recognizer.train(faces, np.array(Ids))
dirName='/home/siva/anaconda3/envs/env/Projects/dbgenerator'
if not os.path.exists(dirName):
    os.mkdir(dirName)
recognizer.save('/home/siva/anaconda3/envs/env/Projects/dbgenerator/dbgenerator.yml')
```

### facial recognition

```python
import cv2
import numpy as np

recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read('/home/siva/anaconda3/envs/env/Projects/dbgenerator/dbgenerator.yml')
faceCascade = cv2.CascadeClassifier('/home/siva/anaconda3/envs/env/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml')
cam = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_SIMPLEX
while True:
    ret, im = cam.read()
    gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray,1.2,5)
    for (x,y,w,h) in faces:
        cv2.rectangle(im,(x,y),(x+w,y+h),(255,0,0),2)
        Id, conf = recognizer.predict(gray[y:y+h,x:x+w])
        if(conf<50):
            if(Id==1):
                Id="Siva"
            elif(Id==2):
                Id="Subramanian"
        else:
            Id="Unknown"
        cv2.putText(im,str(Id),(x,y+h),font,2,255)
    cv2.imshow('im',im)
    k = cv2.waitKey(100) & 0xff
    if k == 27:
        break
cam.release()
cv2.destroyAllWindows()
```
